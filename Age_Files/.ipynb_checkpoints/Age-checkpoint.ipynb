```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv
df = pd.read_csv("Provisional_COVID-19_Death_Counts_by_Sex__Age__and_State.csv")
df
```

```python
# Drop redundant rows to look at only ages
df_rows = df[ (df["Group"] == "By Total") & (df["State"] == "United States") & (df["Sex"] == "All Sexes") ]
df_rows
```

```python
# Drop columns to look at only ages and deaths
df_age = df_rows[["Age Group", "COVID-19 Deaths"]]
df_age
```

```python
# Basic bar graph of deaths by age group including ALL additional groups 
x = df_age["Age Group"]
y = df_age["COVID-19 Deaths"]

plt.bar(x,y)
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
```

```python
# Same as above without the total deaths group
plt.bar(x[1:], y[1:])
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
```

```python
# Line graph of deaths by age group including additional groups (contains overlapping groups, total higher than actual)
plt.plot(x[1:], y[1:], marker='o', color='coral')
plt.title("COVID-19 Deaths by Age Group (All Categories)")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
plt.grid()
```

```python
# Drop additional groups to get continuous groups without overlap
additional = [0, 2, 6, 8, 10, 12]
df_age_cont = df_age.drop(additional)
df_age_cont
```

```python
# Continuous line graph of deaths by age group (The Gem)
x = df_age_cont["Age Group"]
y = df_age_cont["COVID-19 Deaths"]

plt.plot(x,y, marker='o', color='coral')
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
plt.grid()
```

```python
# Cumulative Deaths (way larger than largest number of deaths in any age group, throws off scale)
yvals = y.tolist()
cumu = [0]
for i in range(len(yvals)):
    cumu.append(yvals[i] + cumu[i])
cumu.pop(0)
```

```python
# Continuous and Cumulative not scaled
x = df_age_cont["Age Group"]
y = df_age_cont["COVID-19 Deaths"]

plt.plot(x,cumu, marker='o', color='mistyrose')
plt.plot(x,y, marker='o', color='coral')
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
plt.grid()
```

```python
cumu_percent = [val * 100/sum(yvals) for val in cumu]
```

```python
# Cumulative by percent
x = df_age_cont["Age Group"]
y = df_age_cont["COVID-19 Deaths"]

plt.plot(x,cumu_percent, marker='o', color='mistyrose')
plt.title("COVID-19 Deaths by Age Group (Cumulative, Percent)")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Percent of Total Deaths")
plt.grid()
```

```python
print(cumu_percent)
```

```python
# Scaled to have total death equal to the largest value in the original plot to keep it on the page
cumu_scaled = [val * max(yvals)/sum(yvals) for val in cumu]
```

```python
# Continuous line plot of deaths by age group including cumulative deaths scaled to fit the first continuous graph
x = df_age_cont["Age Group"]
y = df_age_cont["COVID-19 Deaths"]

plt.plot(x,cumu_scaled, marker='o', color='mistyrose')
plt.plot(x,y, marker='o', color='coral')
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
plt.grid()
```

```python
# Bar graph of deaths by age groups, including all additional groups (colored lighter) other than total deaths
x = df_age["Age Group"][1:]
y = df_age["COVID-19 Deaths"][1:]

colors = ['mistyrose' if i in additional else 'coral' for i in df_age.index][1:]

plt.bar(x,y, color=colors)
plt.title("COVID-19 Deaths by Age Group (with Aditional Groups)")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
```

```python
# Overlayed Continuous and Cumulative but a little better
x = df_age_cont["Age Group"]
y = df_age_cont["COVID-19 Deaths"]

fig, ax1 = plt.subplots()
plt.title("COVID-19 Deaths by Age Group (Cumulative, Percent)")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.grid()

ax1.plot(x,y, marker='o', color='coral')
ax1.set_ylabel("Deaths") #, color ='coral'
ax1.tick_params('y') #, colors='coral'

ax2 = ax1.twinx()
ax2.plot(x,cumu_percent, marker='o', color='mistyrose')
ax2.set_ylabel("Percent of Total Deaths") #, color ='mistyrose'

ax1.plot(x,y, marker='o', color='coral')
```

```python
tot = sum(df_age_cont["COVID-19 Deaths"].tolist())
multiplier = [num/tot for num in df_age_cont["COVID-19 Deaths"]]
multiplier
```

```python
# age will be >= bracket
# bracket will be <= age
# Take the biggest bracket your bigger than
brackets = [0, 1, 5, 15, 25, 35, 45, 55, 65, 75, 85]
brackets
```

```python
# 
x = brackets
y = multiplier

plt.plot(x,y, marker='o', color='coral')
plt.title("Vaccine Priority Multiplier by Age Group")
plt.xlabel("Age Group")
#plt.xticks(rotation = 270)
plt.ylabel("Multiplier")
plt.grid()
```

```python
# Takes input age, prints custom bracket, mult, and df category  
test_age = 70
decimals = 5
for r in range(len(brackets)):
    print('index:' , r)
    if test_age < brackets[r]:
        print('bracket:' , brackets[r-1] , ',' , 
              'mult:' , round(multiplier[r-1], decimals) , ',' , 
              'df cat:' , df_age_cont["Age Group"].iloc[r-1])
        break
```
