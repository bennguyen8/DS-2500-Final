```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Provisional_COVID-19_Death_Counts_by_Sex__Age__and_State.csv")
df
```

```python
df_rows = df[ (df["Group"] == "By Total") & (df["State"] == "United States") & (df["Sex"] == "All Sexes") ]
# df_rows = df[ (df["Group"] == "By Total") & (df["State"] == "United States") & (df["Sex"] == "All Sexes") & (df["Age Group"] != "All Ages")]
df_rows
```

```python
df_age = df_rows[["Age Group", "COVID-19 Deaths"]]
df_age
```

```python
x = df_age["Age Group"]
y = df_age["COVID-19 Deaths"]

plt.bar(x,y)
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
```

```python
plt.bar(x[1:], y[1:])
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
```

```python
plt.plot(x[1:], y[1:], marker='o', color='coral')
plt.title("COVID-19 Deaths by Age Group")
plt.xlabel("Age Group")
plt.xticks(rotation = 270)
plt.ylabel("Deaths")
plt.grid()
```
